```python
# PCA for feature extraction
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix, precision_score

# Split data into train and test sets
X = heart_df.drop('output', axis=1)
y = heart_df['output']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.head())

# Apply PCA to the data
pca = PCA(n_components=6)
pca.fit(X_train)

# Transform both the training set and the test set
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
print(X_train_pca.head())

# Create a random forest classifier
rfc = RandomForestClassifier()
rfc.fit(X_train_pca, y_train)

# Evaluate the model
accuracy = rfc.score(X_test_pca, y_test)
print('Accuracy: ', accuracy)
# accuracy = accuracy_score(X_test_pca, y_test)
# print('Accuracy: ', accuracy)

# f1 = f1_score(X_test_pca, y_test)
# print('F1 score: ', f1)

# recall_score = recall_score(X_test_pca, y_test)
# print('Recall score: ', recall_score)

# precision_score = precision_score(X_test_pca, y_test)
# print('Precision score: ', precision_score)

# confusion_matrix = confusion_matrix(X_test_pca, y_test)
# print('Confusion matrix: ', confusion_matrix)
```
